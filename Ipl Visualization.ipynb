```python
wins=df["match_winner"].value_counts()
wins
```

```python
sns.barplot(y=wins.index,x=wins.values,palette="rainbow")
plt.title("most winning Team ")
```

```python
decision_trends=sns.countplot(x=df["toss_decision"],palette="rainbow")
plt.title("decision_trends")
decision_trends
```

```python
toss_vs_wins =df[df["toss_winner"]==df["match_winner"]]["match_id"].count()
percentage=(toss_vs_wins *100)/df.shape[0]
percentage
```

```python
sns.countplot(x=df["won_by"],palette="pastel")
plt.title("how team wins ")
```

```python
df.head()
```

```python
potm=df["player_of_the_match"].value_counts().head(5).sort_values()
sns.barplot(x=potm.values,y=potm.index,palette="Blues")
plt.title("Most Player Of the Match")

```

```python
top_Scorers=df.groupby("top_scorer")["highscore"].sum().sort_values(ascending=False).head(2)
sns.barplot(y=top_Scorers.values,x=top_Scorers.index,palette="Blues")
```

```python
df.head()
```

```python
df["bbf"]=df["best_bowling_figure"].apply(lambda x:x.split("--")[0]).sort_values(ascending=False).astype(int)
Xbbf=df.groupby("best_bowling_figure")["bbf"].max().sort_values(ascending=False).head(10)
Xbbf
```

```python
most_venue=df["venue"].value_counts().sort_values(ascending=False)
sns.barplot(x=most_venue.values,y=most_venue.index,palette="Blues")
```
